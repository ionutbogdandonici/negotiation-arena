```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt

data = pd.read_csv("resource_division_analysis.csv")

plt.figure(figsize=(10, 6))
plt.plot(data["round"], data["fairness_failed_1"], label="Fairness Failed", linewidth=4)
plt.plot(data["round"], data["fairness_reached_1"], label="Fairness Reached (Scenario 1)", linewidth=4)
plt.plot(data["round"], data["fairness_reached_2"], label="Fairness Reached (Scenario 2)", linewidth=4)
plt.plot(data["round"], data["fairness_stopped_1"], label="Fairness Stopped", linewidth=4)
plt.ylim(0, 10)
plt.xlim(1, 10)
plt.grid(True)
plt.legend()
```
